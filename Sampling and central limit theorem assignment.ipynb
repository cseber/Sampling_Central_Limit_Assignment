```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pop1 = np.random.binomial(10, .2, 10000)
pop2 = np.random.binomial(10, .5, 10000)
plt.hist(pop1, alpha=.5, label='population 1')
plt.hist(pop2, alpha=.5, label='population 2')
plt.legend(loc='upper right')
plt.show()
```

```python
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)
plt.hist(sample1, alpha=.6, label='sample 1')
plt.hist(sample2, alpha=.6, label='sample 2')
plt.legend(loc='upper right')
plt.show()
mean_sample1, mean_sample2 = np.mean(sample1), np.mean(sample2)
std_sample1, std_sample2 = np.std(sample1), np.std(sample2)
print('Sample 1 mean: {}'.format(mean_sample1))
print('Sample 2 mean: {}'.format(mean_sample2))
print('Sample 1 std: {}'.format(round(std_sample1, 3)))
print('Sample 2 std: {}'.format(round(std_sample2, 3)))
from scipy.stats import ttest_ind
print(ttest_ind(sample2, sample1, equal_var=False))
```
